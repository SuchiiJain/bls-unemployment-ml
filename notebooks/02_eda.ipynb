```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/processed/unemployment_rate.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
plt.figure()
plt.plot(df["date"], df["unemployment_rate"])
plt.title("U.S. Unemployment Rate Over Time")
plt.xlabel("Year")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```

```python
df["year"] = pd.to_datetime(df["date"]).dt.year

df.groupby("year")["unemployment_rate"].mean().plot()
plt.title("Average Annual Unemployment Rate")
plt.ylabel("Rate (%)")
plt.show()
```

```python
df["month"] = pd.to_datetime(df["date"]).dt.month

monthly_avg = df.groupby("month")["unemployment_rate"].mean()

plt.figure()
plt.plot(monthly_avg.index, monthly_avg.values)
plt.title("Average Unemployment Rate by Month")
plt.xlabel("Month")
plt.ylabel("Rate (%)")
plt.show()
```
