```python
import pandas as pd

df = pd.read_csv("../data/processed/unemployment_rate.csv")
df["date"] = pd.to_datetime(df["date"])
df = df.sort_values("date")

df.head()
```

```python
df["year"] = df["date"].dt.year
df["month"] = df["date"].dt.month
```

```python
df["lag_1"] = df["unemployment_rate"].shift(1)
df["lag_3"] = df["unemployment_rate"].shift(3)
df["lag_6"] = df["unemployment_rate"].shift(6)
df["lag_12"] = df["unemployment_rate"].shift(12)

df.head(15)
```

```python
df["rolling_3"] = df["unemployment_rate"].rolling(3).mean()
df["rolling_6"] = df["unemployment_rate"].rolling(6).mean()
df["rolling_12"] = df["unemployment_rate"].rolling(12).mean()
```

```python
df_fe = df.dropna().reset_index(drop=True)
df_fe.head()
```

```python
features = [
    "year", "month",
    "lag_1", "lag_3", "lag_6", "lag_12",
    "rolling_3", "rolling_6", "rolling_12"
]

X = df_fe[features]
y = df_fe["unemployment_rate"]

X.head(), y.head()
```

```python
df_fe.to_csv("../data/processed/unemployment_features.csv", index=False)
```
